```python
# walrus operator
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")
# It lets you assign a value to a variable as part of an expression, not only as a standalone statement.
if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

```python
# identity operators
# Identity operators are used to compare the objects, 
# not if they are equal, but if they are actually the same object, with the same memory location:

# is operator returns true if both variables are pointing to the same object.

x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
print(x is y)
print(x == y)
```
