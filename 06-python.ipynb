```python
# Decorators let you add extra behavior to a function, without changing the function's code.

# A decorator is a function that takes another function as input and returns a new function.
```

```python
def changecase(func):
    def my_inner():
        return func().upper();
    return my_inner

@changecase
def my_func():
    return "Aman";

print(my_func())
```

```python
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Hello Sally"

@changecase
def otherfunction():
  return "I am speed!"

print(myfunction())
print(otherfunction())
```

```python
def changecase(func):
    def myinner(x):
        return func(x).upper()
    return myinner


@changecase
def my_func(name):
    return f"Hello {name}"
print(my_func("aman"))
```

```python
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

def addgreeting(func):
  def myinner():
    return "Hello " + func() + " Have a good day!"
  return myinner

@changecase
@addgreeting
def myfunction():
  return "Tobias"

print(myfunction())
```

## **generators**

```python
# Generators are functions that can pause and resume their execution.
# When a generator function is called, it returns a generator object, which is an iterator.
# A generator object is a function that can be paused and resumed.
# The code inside the function is not executed yet, it is only compiled. 
# The function only executes when you iterate over the generator.
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
for value in my_generator():
    print(value)
```

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for value in count_up_to(15):
    print(value,end=" , ")
```

```python
# Unlike return, which terminates the function, yield pauses it and can be called multiple times.
```

```python
# Generators are memory-efficient because they generate values on-the-fly 
# instead of storing everything in memory.

# For large datasets, generators save memory:
```

```python
def large_sequence(n):
  for i in range(n):
    yield i

# This doesn't create a million numbers in memory
gen = large_sequence(1000000)
print(next(gen))
print(next(gen))
```

```python
def simple_gen():
  yield "Emil"
  yield "Tobias"
  yield "Linus"

gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def simple_gen():
  yield 1
  yield 2

gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen)) # This will raise StopIteration
```

```python
# generator expression
list_comp = [x*x for x in range(5)]
print(list_comp)
```

```python
# gen-exp
gen_exp = (x*x for x in range(5))
print(gen_exp)
print(list(gen_exp))
```

```python
total = (x * x for x in range(10))
print(sum(total))
```

```python
def fibonacci():
    a,b = 0,1
    while True:
        yield a
        a,b = b,a+b
gen = fibonacci()
for _ in range(10):
    print(next(gen))
```

```python
# The send() method allows you to send a value to the generator:
def gen_():
    while True:
        received = yield
        print("received: ",received)

gen = gen_()
print(next(gen))
gen.send("hello")
gen.send("aman")
gen.send("radha")
```

```python
def my_gen():
  try:
    yield 1
    yield 2
    yield 3
  finally:
    print("Generator closed")

gen = my_gen()
print(next(gen))
print(next(gen))
print(next(gen))
gen.close()
```
