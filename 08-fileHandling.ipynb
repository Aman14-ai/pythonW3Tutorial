```python
with open("demofile.txt","rt") as f:
    print(f.read(),type(f.read()), sep=" || ")
    f.close()
```

```python
with open("demofile.txt","rt") as f:
    print(f.read(15)) # returns the first 5 characters of file
    f.close()
```

```python
# read one line of the file
with open("demofile.txt","rt") as f:
    print(f.readline())
    f.close()
```

```python
with open("demofile.txt","rt") as f:
    for x in f:
        print(x)
    f.close()
```

### **write an existing file**

```python
with open("demofile.txt","a") as f:
    f.write("\nwriting this file");
f = open("demofile.txt","rt")
print(f.read())
f.close()
```

```python
with open("demofile.txt","w") as f:
    f.write("I have deleted the content of the file");
f = open("demofile.txt","rt")
print(f.read())
f.close()
```
